```python
img_path = '/media/yangshun/0008EB70000B0B9F/Datasets/KITTI/data_object_image_2/KITTI/training/image_2'
label_path = '/media/yangshun/0008EB70000B0B9F/Datasets/KITTI/data_object_image_2/KITTI/training/label_2'

imgs = glob.glob(os.path.join(img_path, '*'))
imgs = sorted(imgs)

labels = glob.glob(os.path.join(label_path, '*'))
labels = sorted(labels)

def draw(txt_file, im):
    with open(txt_file, 'rb') as f:
        lines = f.readlines()
    for line in lines:
        lists = line.strip().split(' ')
        cls = lists[0]
        x1 = int(float(lists[4]))
        y1 = int(float(lists[5]))
        x2 = int(float(lists[6]))
        y2 = int(float(lists[7]))
        if cls == 'DontCare':
            cv2.rectangle(im, (x1, y1), (x2, y2), (0, 0, 255), 1)
        else:
            cv2.rectangle(im, (x1, y1), (x2, y2), (0, 255, 0), 2)
            text = str(cls)
            font = cv2.FONT_HERSHEY_SIMPLEX
            cv2.putText(im, text, (x1, y1), font, .4, (0, 0, 255), 1)
    return im

for i in range(len(labels)):
    im = cv2.imread(imgs[i])
    name = os.path.split(imgs[i])[1].split('.')[0]
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(im, name, (5, 35), font, 1, (255, 0, 255), 2)
    
    im = draw(labels[i], im)
    cv2.imshow('o', im)
    
    if cv2.waitKey(0)&0xFF == ord('q'):#按q键退出
        break
    else:
        continue
cv2.destroyAllWindows()
```
